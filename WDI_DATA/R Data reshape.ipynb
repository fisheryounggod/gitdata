```r
# 大数据分块存储
WDI <- import("/Users/mac/Library/CloudStorage/OneDrive-个人/DATA/WDI/WDI-2022.xlsx"，setclass = "data.table")
head(WDI)

# 定义切片的大小
slice_size <- 200

# 循环切片数据并导出为 CSV
for (i in seq(1, ncol(WDI), by = slice_size)) {
  slice <- WDI[, i:min(i + slice_size - 1, ncol(WDI))]
  export(slice, file = paste0("WDI_", i, "_to_", min(i + slice_size - 1, ncol(WDI)), ".csv"))
}
```

```r
# 定义要合并的 CSV 文件的路径
csv_files <- c("WDI.1.csv", "WDI.2.csv", "WDI.3.csv", "WDI.4.csv", "WDI.5.csv", "WDI.6.csv", "WDI.7.csv")

# 定义一个空的数据框
combined_dt <- data.table()

# 使用循环读取 CSV 文件
for (file in csv_files) {
  # 读取 CSV 文件
  dt <- fread(file)

  # 将数据框追加到 combined_dt
  combined_dt <- cbind(combined_dt, dt)
}

# 查看合并后的数据框
combined_dt
```
